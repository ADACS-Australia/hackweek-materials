```python
import matplotlib.pyplot as plt

from astropy.wcs import WCS
from astropy.io import fits
```

```python
# URL for an image.
url = 'https://storage.googleapis.com/panoptes-survey/PAN001/M81/14d3bd/20191108T144854/20191108T150207.fits.fz'

# Get the data, header, and WCS.
data0 = fits.getdata(url)
header0 = fits.getheader(url, ext=1)
wcs0 = WCS(header0)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1, projection=wcs0)
ax.imshow(data0);
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1, projection=wcs0)
im = ax.imshow(data0)

# Testing out the grid plot.
ax.coords[0].set_ticklabel_position('l')
ax.coords[0].grid(color='white', ls='dotted')
ax.set_ylabel('RA')
ax.coords[1].set_ticklabel_position('b')
ax.coords[1].grid(color='white', ls='dotted')
ax.set_xlabel('Dec');
```
